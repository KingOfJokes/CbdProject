```python
import sklearn as skl
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import math
import numpy as np
import pickle
import random
import time
import os
```

```python
Xs = os.listdir('./bt/btxa')
Ys = os.listdir('./bt/bty')
```

```python

#for i in range(len(Xs)):
    #print(Xs[i],Ys[i])
for i in range(len(Xs)):
    print(Xs[i],Ys[i])
    x_list = np.load('./bt/btxa/'+Xs[i],allow_pickle = True)
    x_list = x_list.tolist()
    x_array = []
    for j in range(len(x_list)):
        temp = []
        for key in x_list[j].keys():
            temp.append(x_list[j][key])
        x_array.append(temp)

    for v in range(len(x_array)):
        n = np.asarray(x_array[v])
        if np.sum(n * n) == 0:
            pass
        else:
            x_array[v] = n / (np.sum(n * n) ** 0.5)
            x_array[v] = x_array[v].tolist()
            #x_array
    
    y = np.load('./bt/bty/'+Ys[i],allow_pickle = True)
    y_array = y.tolist()
    
    print('#測資',len(x_array))
    print('#字詞',len(x_array[0]))
    
    if len(x_array) == 0:
        print('empty')
    else:
        x_train = x_array
        y_train = y_array

        model = RandomForestClassifier(n_estimators = 100)

        model.fit(x_train,y_train)
        output = model.predict(x_train)
        y_pred = output.tolist()
        accuracy = np.mean(output == y_train)
        print(accuracy)
                #print(output)
                #print(y_pred)
                #print(y_test)
                #output
                #model
        con = m.confusion_matrix(y_pred,y_train)
        print(con)
        name = Xs[i].split('atr')[0]
        with open('./bt/model/'+name+'model.pickle','wb') as f:
            pickle.dump(model,f)
```
